# Matplotlib

```
import numpy as np
import matplotlib.pyplot as plt
```

```
a = np.array(list(range(10)))
a
```

```
plt.plot(a, a)
plt.show()
```

```
plt.scatter(a, a)
plt.show()
```

```
plt.plot(a[:5], a[:5], c = 'r')
plt.plot(a[:5], (a**2)[:5], c = 'b')
plt.plot(a[:5], (a**3)[:5], c = 'g')
plt.axis([0, 5, 0, 5])
plt.show()
```

```
a = np.arange(0, 10, 0.01)
# print(a)
plt.plot(a, a, c = 'r')
plt.plot(a, (a**2), c = 'b')
plt.plot(a, (a**3), c = 'g')
plt.axis([0, 10, 0, 10])
plt.grid()
plt.xlabel("xyz")
plt.ylabel("abc")
plt.show()
```

```
a = np.random.randint(0, 10, 10)
c = np.random.randint(0, 10, 10)
b = np.array(list(range(10)))
a, b
```

```
plt.bar(b + 0.25, a, width= 0.25)
plt.bar(b, c, width= 0.25)
plt.show()
```

### pie chart

```
a = ["apple", "banana", 'magno', 'pineapple']
b = np.random.randint(80, 200, 4)
a, b
```

```
plt.pie(b, labels= a, explode= [0.05, 0, 0, 0.15])
plt.show()
```
